## Generators
Simply, generators are iterators that you can only iterate over once. To understand generators, we need to know some things about iterators. 

An iterator is an object that let's a programmer traverse a container, mainly lists.

Parts of iterators in Python:
- Iterable
An iterable is any object in Python with a `__iter__` method or `__getitem__` method by which it can take indexes. Iterable provides us with iterator.

- Iterator
Iterator is an object with `next()` or `__next__` method defined.

- Iteraton
Iteration is a process of looping over a iterator.


### Generators
Generators are also iterators in Python, but we can only iterate over them once. It's because they do not store all the values in the memory, they generate value on the fly, and we can use them by iterating over them. Either with `for` loop or by passing to any function or construct that can do the iteration. However, it doesn't return the value of function. It yields it.

"Syntactically, a generator is a function with a yield statement."

>>difference between return and yield: 

>> **yield**: it suspends the execution of function, returns value to caller and saves the state.

>> **return** : it ends the execution of function, and returns value to the caller.


A simple example of generator function:

```python
def gen_func():
    for i in range(10):
        yield i
```

```python
# Calling the generator with the iterator
for item in gen_func():
    print(item)
```

The purpose of the generator is not obvious in this example and generator isn't that useful here. It is useful in cases where the state of the exectuion is important. Especially, if we don't want to allocate the memory for all results at the same time.

**think of a generator generating a prime number or fibonacci series when you call it.**

```python
#fibonacci generator
def fibonacci(n):
    a = b = 1
    for i in range(n):
        yield a
        a, b = b, a + b
```

```python
# nowe can use it whenever we like
for numbers in fibonacci(5):
    print(numbers)
```

```python
for nums in fibonacci(6):
    print(nums)
```
